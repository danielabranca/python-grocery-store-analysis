#### Question:
Determine a suitable way to combine the orders_products_combined dataframe with your products data set. Make sure you’re using your wrangled, cleaned, and deduped products data set stored in your “Prepared Data” folder from the previous Exercise’s task.

Confirm the results of the merge using the merge flag.

#### Answer:
The data frames have the column "product_id" in common. An inner join can be used.

```python
# Testing Merging without overwriting anything (with the help of Richard): 
# function merge: (dataframes to merge, 'on' means which columns are the key, 'how' means what kind of merge, 
# 'indicator' creates a new column to check how the df matched.
pd.merge(df_ords_prods, df_prods, on = ['product_id'], how='inner', indicator='match')
```

```python
# merging the two df: merging df_ords_prods with df_prods, with 'product_id' as key, keeping track on if each row is matching or not.

ords_prods_merge = df_ords_prods.merge(df_prods , on = ['product_id'] , indicator = 'match' )
```

```python
# checking how the match is looking:
ords_prods_merge['match'].value_counts()
```
